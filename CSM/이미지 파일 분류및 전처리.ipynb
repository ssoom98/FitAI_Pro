```python
import os
import re
import shutil

# 최상위 디렉터리 경로 설정
base_directory = r'E:\korean_shape\data'

# 원하는 파일 이름 형식에 맞는 정규식 패턴
file_pattern = re.compile(r'^\d{2}_\d{2}_[A-Z]\d{3}_\d{2}\.jpg$')

# 필터링 기준 설정
keep_poses = ['02']              # 유지할 자세 번호: 02번
keep_clothing = ['01','03']           # 유지할 복장 번호: 03번
keep_models = [f'M{str(i).zfill(3)}' for i in range(9, 505)]  # F009 ~ F504 모델
keep_angles = ['04', '08', '12', '16', '20', '24', '28', '32']  # 유지할 각도 번호들

# 파일 유지 여부를 판단하는 함수
def should_keep_file(filename):
    match = file_pattern.match(filename)
    if not match:
        return False

    # 파일명에서 요소 추출
    pose, clothing, model, angle = filename.split('_')
    angle = angle.split('.')[0]  # 확장자 제거

    # 조건에 맞는지 확인
    return (
        pose in keep_poses and
        clothing in keep_clothing and
        model in keep_models and
        angle in keep_angles
    )

# 파일을 처리하는 함수
def process_model_folder(model_folder):
    image_folder = os.path.join(model_folder, 'Image')
    move_folder = os.path.join(model_folder, 'use_image')

    # 이동할 폴더가 없으면 생성
    if not os.path.exists(move_folder):
        os.makedirs(move_folder)

    # Image 폴더 내 파일 처리
    for filename in os.listdir(image_folder):
        if should_keep_file(filename):
            source_path = os.path.join(image_folder, filename)
            destination_path = os.path.join(move_folder, filename)
            print(f"이동할 파일: {source_path} -> {destination_path}")
            shutil.move(source_path, destination_path)

# 모델별 폴더 순회 및 처리
def process_all_models():
    for model_name in os.listdir(base_directory):
        model_folder = os.path.join(base_directory, model_name)
        if os.path.isdir(model_folder) and model_name in keep_models:
            process_model_folder(model_folder)

# 실행
process_all_models()
```

```python
import os
import pandas as pd

# 데이터 디렉토리 경로 및 필요한 컬럼명 매핑
data_dir = r"E:\151.한국인 전신 형상 및 치수 측정 데이터\01.데이터\1.Training\원천데이터\신체정보데이터"
columns_map = {
    'Unnamed: 2': '모델번호',
    '키': '키',
    '몸무게': '몸무게',
    '체지방율': '체지방율',
    '허리둘레': '허리둘레',
    '배꼽수준허리둘레':'배꼽수준허리둘레',
    '엉덩이둘레': '엉덩이둘레',
    '넙다리둘레': '넙다리둘레',
    '젖가슴둘레': '젖가슴둘레',
    '편위팔둘레': '편위팔둘레',
    '어깨사이너비': '어깨사이너비',
    '샅높이': '샅높이',
    '성별': '성별',
    '나이': '나이'
}

# 데이터 저장을 위한 빈 리스트
data_records = []

# 데이터 처리
for root, dirs, files in os.walk(data_dir):
    for file in files:
        if file.endswith(".csv"):
            file_path = os.path.join(root, file)
            try:
                # CSV 파일 읽기
                df = pd.read_csv(file_path, encoding='utf8', header=1)
                
                # 필요한 컬럼만 추출 및 컬럼명 변경
                if set(columns_map.keys()).issubset(df.columns):
                    filtered_data = df.rename(columns=columns_map)[columns_map.values()]
                    data_records.append(filtered_data)
                else:
                    print(f"필요한 컬럼이 없는 파일: {file_path}")
            except Exception as e:
                print(f"파일 처리 중 오류 발생: {file_path}, 오류: {e}")

# 데이터프레임으로 합치기
if data_records:
    final_data = pd.concat(data_records, ignore_index=True)
    
    # 결과를 CSV 파일로 저장
    output_file = r"E:\통합_신체정보데이터.csv"
    final_data.to_csv(output_file, index=False, encoding='utf-8-sig')
    print(f"통합 파일이 성공적으로 저장되었습니다: {output_file}")
else:
    print("데이터가 수집되지 않았습니다. 경로 및 파일 상태를 확인하세요.")
```

```python
import cv2
import glob
import os
import numpy as np
from ultralytics import YOLO

# YOLO 모델 로드
model_name = 'yolov8n'  # YOLO 모델 이름
model = YOLO(f'{model_name}.pt')

# 저장할 폴더 경로 설정
save_dir = "E:/image_data"
os.makedirs(save_dir, exist_ok=True)  # 폴더가 없으면 생성

# 화면 해상도 설정
screen_width = 1920  # 모니터 가로 크기
screen_height = 1080  # 모니터 세로 크기

# 모델 코드 생성 (F009 ~ F504, M009 ~ M505)
female_models = [f"F{str(i).zfill(3)}" for i in range(9, 505)]
male_models = [f"M{str(i).zfill(3)}" for i in range(9, 506)]
model_list = female_models + male_models  # 전체 모델 리스트

# 모델별 이미지 처리
for model_code in model_list:
    image_paths = glob.glob(f'E:/korean_shape/data/{model_code}/use_image/*.jpg')
    
    save_index = 1  # 저장할 이미지 번호 (1~16)
    
    for image_path in image_paths:
        # 이미지 로드
        img = cv2.imread(image_path)
        
        # 원본 이미지 파일명 가져오기 (확장자 제외)
        base_filename = os.path.splitext(os.path.basename(image_path))[0]

        # YOLO 실행 (탐지 수행)
        results = model(image_path)

        for result in results:
            boxes = result.boxes  # 탐지된 박스 정보 가져오기

            for i, box in enumerate(boxes):
                x1, y1, x2, y2 = map(int, box.xyxy[0])  # 좌상단(x1, y1), 우하단(x2, y2) 좌표
                confidence = box.conf[0].item()  # 신뢰도
                class_id = int(box.cls[0].item())  # 클래스 ID

                # 사람(클래스 ID 0)만 처리
                if class_id == 0:
                    # 탐지된 부분 크롭 (자르기)
                    cropped_img = img[y1:y2, x1:x2]

                    # 크롭한 이미지가 너무 작으면 건너뜀
                    if cropped_img.shape[0] < 10 or cropped_img.shape[1] < 10:
                        continue

                    # 원본 크기 가져오기
                    h, w = cropped_img.shape[:2]

                    # 256x256 배경(검은색) 생성
                    new_img = np.zeros((256, 256, 3), dtype=np.uint8)

                    # 비율 유지하면서 크기 조정
                    if h > w:
                        scale = 256 / h
                        new_w = int(w * scale)
                        resized_img = cv2.resize(cropped_img, (new_w, 256))  # 높이를 256으로 맞춤
                        x_offset = (256 - new_w) // 2  # 중앙 정렬
                        new_img[:, x_offset:x_offset + new_w] = resized_img  # 중앙에 배치
                    else:
                        scale = 256 / w
                        new_h = int(h * scale)
                        resized_img = cv2.resize(cropped_img, (256, new_h))  # 너비를 256으로 맞춤
                        y_offset = (256 - new_h) // 2  # 중앙 정렬
                        new_img[y_offset:y_offset + new_h, :] = resized_img  # 중앙에 배치

                    # 저장할 파일 이름 설정 (resize_모델번호_1~16.jpg)
                    save_path = os.path.join(save_dir, f"resize_{model_code}_{save_index}.jpg")
                    cv2.imwrite(save_path, new_img)  # 이미지 저장
                    print(f"[저장 완료] {save_path}")

                    # 저장한 이미지 번호 증가
                    save_index += 1
                    if save_index > 16:
                        break  # 16장 저장 후 종료

        if save_index > 16:
            break  # 16장 저장 후 종료

print("이미지 저장 완료")
```
