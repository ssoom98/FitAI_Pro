```python
import pandas as pd
df = pd.read_csv('CNN_특징벡터_신체데이터.csv')
df.head()
```

```python
bodyshape = pd.read_csv("E://통합_신체정보데이터.csv")
bodyshape.head()
```

```python
df_merged = df.merge(bodyshape, on = "model_num",how="inner")
df_merged
```

```python
X = df_merged.drop(columns=['model_num', 'body_shape'])  # EfficientNetB0 + VGG16 특징 벡터
y = df_merged['body_shape']   # 체형 분류

y
```

```python
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
from sklearn.preprocessing import LabelEncoder
from imblearn.over_sampling import SMOTE
from imblearn.over_sampling import ADASYN
from sklearn.utils.class_weight import compute_sample_weight
import numpy as np
import pandas as pd

# 라벨 인코딩 (한 번만 수행)
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# 범주형 변수 변환 (성별 컬럼 인코딩)
X['gender'] = LabelEncoder().fit_transform(X['gender'])

# 데이터 분할 (층화 유지)
X_train, X_test, y_train, y_test = train_test_split(
    X, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded
)
sample_weights = compute_sample_weight(class_weight='balanced', y=y_train_resampled)
# 클래스별 샘플 개수 확인
unique, counts = np.unique(y_train, return_counts=True)
class_counts = dict(zip(unique, counts))
print("클래스별 샘플 개수:", class_counts)

# SMOTE 적용 (소수 클래스만 오버샘플링)
adasyn = ADASYN(sampling_strategy={5: 200}, random_state=42)
X_train_resampled, y_train_resampled = adasyn.fit_resample(X_train, y_train)
# XGBoost 분류 모델 생성
model = xgb.XGBClassifier(
    n_estimators=500,  
    learning_rate=0.07,  
    max_depth=5,  
    min_child_weight=3,  
    gamma=0.1,  # 기존 0.2 → 0.1로 완화 (과적합 방지)
    max_delta_step=1,  # 가중치를 더 균형 있게 업데이트하도록 조정
    subsample=0.8,  
    colsample_bytree=0.8,  
    reg_alpha=0.1,  
    reg_lambda=0.5,  
    objective="multi:softmax",
    num_class=len(label_encoder.classes_),
    eval_metric="mlogloss",
    random_state=42
)

# 모델 학습
model.fit(X_train_resampled, y_train_resampled, sample_weight=sample_weights)

# 예측 수행
y_pred = model.predict(X_test)

# 평가 지표 출력
accuracy = accuracy_score(y_test, y_pred)
print(f"분류 정확도 (Accuracy): {accuracy:.4f}")

print("분류 리포트:")
print(classification_report(y_test, y_pred, target_names=label_encoder.classes_, zero_division=1))

# 예측 결과 비교 (DataFrame)
df_results = pd.DataFrame({
    'Actual': label_encoder.inverse_transform(y_test),
    'Predicted': label_encoder.inverse_transform(y_pred)
})
print(df_results.head())
```
