```python
import pandas as pd
df = pd.read_csv('data/CNN_특징벡터_신체데이터.csv')
df.head()
```

```python
bodyshape = pd.read_csv("E://통합_신체정보데이터.csv")
bodyshape.head()
```

```python
df_merged = df.merge(bodyshape, on = "model_num",how="inner")
df_merged
```

```python
import numpy as np
X = df_merged.drop(columns=['model_num', 'body_shape'])  # EfficientNetB0 + VGG16 특징 벡터
y = df_merged['body_shape']   # 체형 분류

unique, counts = np.unique(y, return_counts=True)
class_counts = dict(zip(unique, counts))
print("클래스별 샘플 개수:\n", class_counts)
```

```python
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
from sklearn.preprocessing import LabelEncoder
from imblearn.over_sampling import ADASYN
from sklearn.utils.class_weight import compute_sample_weight
import numpy as np
import pandas as pd

# 데이터셋 로드 및 전처리
# X, y는 기존 데이터프레임 (사용자가 정의했다고 가정)
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# 범주형 변수 변환 (성별 컬럼 인코딩)
X['gender'] = LabelEncoder().fit_transform(X['gender'])

# 데이터 분할 (층화 유지)
X_train, X_test, y_train, y_test = train_test_split(
    X, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded
)

# ADASYN 적용 (소수 클래스만 오버샘플링)
adasyn = ADASYN(sampling_strategy="auto", random_state=42)
X_train_resampled, y_train_resampled = adasyn.fit_resample(X_train, y_train)

# 샘플 가중치 계산
sample_weights = compute_sample_weight(class_weight='balanced', y=y_train_resampled)

# 클래스별 샘플 개수 확인
unique, counts = np.unique(y_train_resampled, return_counts=True)
class_counts = dict(zip(unique, counts))
print("클래스별 샘플 개수:", class_counts)

# XGBoost 분류 모델 생성
model = xgb.XGBClassifier(
    n_estimators=500,  
    learning_rate=0.07,  
    max_depth=5,  
    min_child_weight=3,  
    gamma=0.1,  # 기존 0.2 → 0.1로 완화 (과적합 방지)
    max_delta_step=1,  # 가중치를 더 균형 있게 업데이트하도록 조정
    subsample=0.8,  
    colsample_bytree=0.8,  
    reg_alpha=0.1,  
    reg_lambda=0.5,  
    objective="multi:softmax",
    num_class=len(label_encoder.classes_),
    eval_metric="mlogloss",
    random_state=42
)

# 모델 학습
model.fit(X_train_resampled, y_train_resampled, sample_weight=sample_weights)

```

```python
# 예측 수행
y_pred = model.predict(X_test)

# 평가 지표 출력
accuracy = accuracy_score(y_test, y_pred)
print(f"분류 정확도 (Accuracy): {accuracy:.4f}")

# 예측 결과 비교 (DataFrame)
df_results = pd.DataFrame({
    'Actual': label_encoder.inverse_transform(y_test),
    'Predicted': label_encoder.inverse_transform(y_pred)
})
print(df_results.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import f1_score, confusion_matrix
import warnings
# 한글 폰트 설정 (matplotlib에서 한글 깨짐 방지)
import matplotlib.font_manager as fm

plt.rc('font', family='Malgun Gothic')  # Windows 사용자의 경우
plt.rcParams['axes.unicode_minus'] = False  # 마이너스 부호 깨짐 방지

# 모든 경고 메시지 무시
warnings.filterwarnings("ignore")
# 분류 리포트 생성
report = classification_report(y_test, y_pred, target_names=label_encoder.classes_, zero_division=1, output_dict=True)

# DataFrame 변환 (깔끔한 출력용)
df_report = pd.DataFrame(report).T

# 불필요한 행(Overall Accuracy, macro avg, weighted avg) 제거
df_report = df_report.drop(['accuracy', 'macro avg', 'weighted avg'], errors='ignore')

# 시각적으로 보기 좋게 변환
df_report = df_report[['precision', 'recall', 'f1-score', 'support']]

# 분류 리포트 출력 (표 형태)
print("\n=== 분류 리포트 ===")
print(df_report.round(4).to_string())

# 분류 리포트 시각화 (파란색 계열 바 그래프)
plt.figure(figsize=(12, 6))
df_report.iloc[:, :3].plot(kind='bar', figsize=(12, 6), color=['#1f77b4', '#6baed6', '#9ecae1'])  # 파란색 계열 적용
plt.title("Precision, Recall, F1-score 비교", fontsize=14)
plt.xlabel("클래스", fontsize=12)
plt.ylabel("Score", fontsize=12)
plt.xticks(rotation=45)
plt.legend(loc='lower right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# F1 스코어 계산 (macro, weighted 평균)
f1_macro = f1_score(y_test, y_pred, average='macro')
f1_weighted = f1_score(y_test, y_pred, average='weighted')

# F1-score 시각화를 위한 데이터프레임 생성
f1_scores = pd.DataFrame({
    "Score Type": ["Macro F1-score", "Weighted F1-score"],
    "Score": [f1_macro, f1_weighted]
})
# 혼동 행렬 생성
conf_matrix = confusion_matrix(y_test, y_pred)

# 시각화 설정
fig, axes = plt.subplots(1, 2, figsize=(15, 6), gridspec_kw={'width_ratios': [1, 2]})

# F1-score 바 그래프 (왼쪽)
sns.barplot(x="Score Type", y="Score", data=f1_scores, ax=axes[0], palette="Blues_r")
axes[0].set_ylim(0, 1)
axes[0].set_ylabel("F1 Score")
axes[0].set_title("F1-score 비교")

# 혼동 행렬 히트맵 (오른쪽)
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=label_encoder.classes_, yticklabels=label_encoder.classes_, ax=axes[1])
axes[1].set_xlabel('예측값')
axes[1].set_ylabel('실제값')
axes[1].set_title('혼동 행렬 (Confusion Matrix)')

# 그래프 출력
plt.tight_layout()
plt.show()

```
